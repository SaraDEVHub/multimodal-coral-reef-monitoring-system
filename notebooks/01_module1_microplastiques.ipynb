# Module 1 — Détection des microplastiques

Ce notebook vérifie d'abord la structure du projet et la présence des données.

```python
from pathlib import Path
import pandas as pd

# Fonction robuste : fonctionne si Jupyter est lancé depuis hind/ ou hind/notebooks/
cwd = Path.cwd()
if (cwd / 'data').exists():
    PROJECT_ROOT = cwd
elif (cwd.parent / 'data').exists():
    PROJECT_ROOT = cwd.parent
else:
    raise FileNotFoundError(
        "Impossible de trouver le dossier 'data'. "
        "Lance Jupyter depuis le dossier 'hind' ou vérifie l'organisation du projet."
    )

MODULE1_DIR = PROJECT_ROOT / 'data' / 'module1'
MODULE2_FILE = PROJECT_ROOT / 'data' / 'module2' / 'HAB_Artificial_GAN_Dataset.csv'
MODULE3_CSV = PROJECT_ROOT / 'data' / 'module3' / 'ReefNet_RSG-TestDataset_annotations.csv'
MODULE3_IMAGES = PROJECT_ROOT / 'data' / 'module3' / 'ReefNet_RSG-TestDataset_images'

module1_files = sorted(MODULE1_DIR.glob('*.csv'))

print('Racine du projet :', PROJECT_ROOT)
print('Dossier module 1 :', MODULE1_DIR)
print('Nombre de CSV Module 1 :', len(module1_files))
print('Module 2 trouvé :', MODULE2_FILE.exists())
print('Annotations Module 3 trouvées :', MODULE3_CSV.exists())
print('Dossier images Module 3 trouvé :', MODULE3_IMAGES.exists())

if len(module1_files) == 0:
    raise FileNotFoundError(f'Aucun fichier CSV trouvé dans {MODULE1_DIR}')

print('\nPremiers fichiers du module 1 :')
for file in module1_files[:5]:
    print('-', file.name)

print('\n✅ Vérification terminée.')
```
